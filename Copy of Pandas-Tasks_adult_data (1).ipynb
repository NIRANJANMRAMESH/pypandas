Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
conditions = [
    (data['age'] <= 30),
    (data['age'] > 30) & (data['age'] <= 60),
    (data['age'] > 60)
    ]
values = ['young', 'adult', 'senior']

data['age_group'] = np.select(conditions, values)

data.head()
```
